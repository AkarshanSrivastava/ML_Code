# loc vs. iloc

https://www.analyticsvidhya.com/blog/2020/02/loc-iloc-pandas/

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    'age' :     [ 10, 22, 13, 21, 12, 11, 17],
    'section' : [ 1, 2, 1, 1, 2, 2, 1],
    'city' :    [ 'Gurgaon', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai'],
    'gender' :  [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'favourite_color' : [ 'red', np.NAN, 'yellow', np.NAN, 'black', 'green', 'red']
})
```

```python
data
```

```python
data
```

```python
data.iloc[1:,0:3]
```

```python
data.iloc[1]=+10
```

```python
data
```

```python
data.iloc[0,0]+=10
data
```

```python
data.loc[1,"age"]+=10
data
```

```python
data.loc[data.age>15]
```

```python
data[data.age>15][["age","city"]]
```

```python
data.loc[(data.age>15),["gender"]]='M'
data
```

```python
data.loc[(data.age>12),["city",'section']]=("pune","c")
data
```

```python
data.loc[(data.age>30) &(data.city=="pune")]
```
