```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X= iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search=GridSearchCV(estimator=AdaBoostClassifier(),
             param_grid={'base_estimator': ('classes_','n_classes_',None ),
                         'n_estimators': [50,100,20,150],
                        'learning_rate':[1,2,5,3],
                        'algorithm':('SAMME', 'SAMME.R'),},
                        cv=5)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
model = AdaBoostClassifier(n_estimators=150,algorithm='SAMME',
                         learning_rate=1)
```

```python
# Train Adaboost Classifer
model = model.fit(X_train, y_train)

```

```python
#Predict the response for test dataset
y_pred_train = model.predict(X_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_train, y_pred_train)
```

```python
metrics.accuracy_score(y_test, y_pred_test)
```

```python
#diff base learner 
# Import Support Vector Classifier
from sklearn.svm import SVC
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
svc=SVC(probability=True, kernel='linear')

# Create adaboost classifer object
abc =AdaBoostClassifier(n_estimators=50, base_estimator=svc,learning_rate=1)

# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(model, X_train, y_train, cv=10)
```

```python
score.mean()
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
train_predict=cross_val_predict(model,X_train,y_train,cv=4)
```

```python
print("Accuracy:",metrics.accuracy_score(y_train, train_predict))
```

```python
predict=cross_val_predict(model,X_test,y_test,cv=4)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, predict))
```
